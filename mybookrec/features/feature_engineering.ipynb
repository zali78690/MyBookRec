```python
import json
import polars as pl
from mybookrec import DATA_DIR, ROOT_DIR
```

```python
# Read genres column from transformed books parquet file and extract unique genres
genres_struct = pl.read_parquet(DATA_DIR / "transformed" / "books_with_genres.parquet").schema["genres"]

# Extract genres into normal python list
unique_genres = []
for genre in genres_struct.fields:
    unique_genres.append(genre.name)

print(f"Unique genres:", unique_genres)
```

```python
# Dump unique genres to json file
with open(ROOT_DIR / "mybookrec"/ "features"/ "genre_vocab.json", "w") as file:
    json.dump(unique_genres, file, indent=2)
```
